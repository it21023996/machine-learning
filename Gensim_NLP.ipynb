```python
import pandas as pd
```

```python
df= pd.read_csv("Book1.csv", on_bad_lines='skip')
```

```python
df.iloc[2998:3004]
```

```python
df.shape
```

```python
df.Label.value_counts()
```

```python
df['new_label'] = df['Label'].apply(lambda x:1 if x=='TRUE' else 0)
```

```python
df.iloc[2996:3004]
```

```python
!python -m spacy download en_core_web_lg
```

```python
import spacy
nlp = spacy.load('en_core_web_lg')
```

```python
doc = nlp('top trump surrogate brutally stabs his')
```

```python
doc.vector.shape
```

```python
df['vectors'] = df['Text'].apply(lambda text:nlp(text).vector)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.vectors.values,df.new_label,test_size=0.2,random_state=2022)
```

```python
x_train
```

```python
import numpy as np
```

```python
x_train_2d = np.stack(x_train)
x_test_2d = np.stack(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
x_train_2d = scaler.fit_transform(x_train_2d)
x_test_2d = scaler.transform(x_test_2d)
```

```python
clf = MultinomialNB()
```

```python
clf.fit(x_train_2d,y_train)
```

```python
y_pred = clf.predict(x_test_2d)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```
